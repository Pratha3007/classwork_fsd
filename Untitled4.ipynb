```python
def grade(x):
    if(x>300):
        return "A+"
    elif(x>250):
        return "A"
    elif(x>200):
        return "B+"
    return 'F'
```

```python
df["Grade"]=df["Total"].apply(grade)
```

# Using Lambda

```python
df["Grade"]=df.apply(lambda x:"Fail" if (x["A"]<10) or (x["B"]<10) or (x["C"]<10) or (x["D"]<10) or (x["E"]<10) else "Pass" ,axis=1)
```

```python
df
```

# using user defined function

```python
def result(x):
    if (x["A"]<10) or (x["B"]<10) or (x["C"]<10) or (x["D"]<10) or (x["E"]<10):
        return "Fail"
    return "Pass"
```

```python
df["Result"]=df.apply(result,axis=1)
```
