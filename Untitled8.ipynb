# Concat and Merge

```python
import numpy as np
```

```python
import pandas as pd
```

```python
df1=pd.DataFrame({"A":[1,2],'B':[3,4]})
df2=pd.DataFrame({"A":[5,6],'B':[7,8]})
```

```python
df1
```

```python
df2
```

```python
pd.concat((df1,df2),ignore_index=True)# axis=0 atle row wise, ignore index=true atle old index ne ignore krine navi index ape
```

```python
pd.concat((df1,df2),axis=1,ignore_index=True) ## axis=1 atle cols wise
```

```python
pd.concat((df1,df2),ignore_index=False,keys=["df1","df2"])
```

```python
df1=pd.DataFrame({"A":[1,2],'B':[3,4]})
df2=pd.DataFrame({"C":[5,6],'B':[7,8]})

df1
```

```python
df2
```

```python
pd.concat((df1,df2),ignore_index=True)
```

```python
pd.concat((df1,df2),axis=0,sort=True,join="inner")# common column ,by default:outer
```

```python
df1=pd.DataFrame({"D":[1,2],'B':[3,4]})
df2=pd.DataFrame({"C":[5,6],'A':[7,8]})

df1
```

```python
pd.concat((df1,df2),sort=True)# column sort kre
```

```python
df1=pd.DataFrame({"name":["Pratha","Naitri","Yug","Mann","Jash"],'Subject':["Python","Fsd","Coa","Dm","Toc"],"id":[3,2,1,4,6]})
df2=pd.DataFrame({"id":[1,2,3,4,5],"Marks":[1,4,2,4,7]})

df1
```

```python
df2
```

```python
pd.merge(df1,df2,on="id",how="inner")
```

```python
pd.merge(df1,df2,on="id",how="outer")
```

```python
pd.merge(df1,df2,on="id",how="left")
```

```python
pd.merge(df1,df2,on="id",how="right")
```

```python
pd.merge(df1,df2,how="cross")
```

```python
pd.merge(df1,df2,left_index=True,right_index=True,suffixes=["_df1","_df2"])
```

```python
df1=pd.DataFrame({"name":["Pratha","Naitri","Yug","Mann","Jash"],'Subject':["Python","Fsd","Coa","Dm","Toc"],"in":[3,2,1,4,6]})
df2=pd.DataFrame({"id":[1,2,3,4,5],"Marks":[1,4,2,4,7]})

df1
```

```python
df1.merge(df2,left_on="in",right_on="id")
```

```python
df1=pd.DataFrame({"name":["Pratha","Naitri","Yug","Mann","Jash"],'Subject':["Python","Fsd","Coa","Dm","Toc"],"id":[3,2,1,4,6]})
df2=pd.DataFrame({"id":[1,2,3,4,5],"Marks":[1,4,2,4,7],"Fac":["Pratha","Naitri","Yug","Mann1","Jash"]})
```

```python
df1.merge(df2,left_on="name",right_on="Fac",how="outer")
```
