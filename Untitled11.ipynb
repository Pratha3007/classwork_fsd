```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.io as pio
pio.renderers.default="notebook"
```

```python
df=pd.DataFrame({"dep":["hr","it","hr","it","hr"],"age":[20,25,30,22,10],"sal":[100,500,200,300,100],"sex":["F","M","M","F","F"]})
```

```python
df
```

```python
px.histogram(data_frame=df,x="age",color="sex")
```

```python
px.histogram(data_frame=df,x="age",color="sex",nbins=15,pattern_shape="dep")
```

```python
px.histogram(data_frame=df,x="dep",y="sal")
```

```python
px.histogram(data_frame=df,x="dep",y="sal",histfunc="max")
```

```python
px.histogram(data_frame=df,x="dep",y="sal",histfunc="min")# x,y both complusory to give
```

```python
px.histogram(data_frame=df,x="dep",y="sal",histfunc="avg")
```

```python
px.histogram(data_frame=df,x="dep",y="sal",histfunc="count")
```
