# Social media csv

```python
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("social_media_engagement.csv")
```

```python
df
```

```python
sns.heatmap(df.corr(numeric_only=True),cmap="coolwarm")
```

```python
sns.heatmap(df.corr(numeric_only=True),cmap="coolwarm",linecolor="white",linewidths=4,annot=True) # under numbers dekhase
```

```python
sns.heatmap(df.corr(numeric_only=True),cmap="coolwarm",linecolor="white",linewidths=4,annot=True,cbar=False) # side ni color line nikdi jase
```

```python
sns.heatmap(df.corr(numeric_only=True),cmap="coolwarm",linecolor="white",linewidths=4,annot=True,vmin=0.2)# 0.2 ni  niche j value hase te badhi same thy jase
```

```python
sns.boxplot(data=df,x='likes')
```

```python
sns.boxplot(data=df,x='likes', hue="platform" ,color="red")
```

```python
sns.boxplot(data=df,x='likes',hue="post_type")
```

```python
sns.scatterplot(data=df,x="likes",y="comments",hue="platform",style="platform",s=100,palette=["red","yellow","blue"])
```
