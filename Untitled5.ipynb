```python
import pandas as pd
```

```python
import numpy as np
```

```python
df=pd.read_csv("social_media_engagement.csv")
```

```python
df
```

```python
df["Quality"]=df.apply(lambda x: " Good " if (x["likes"]>3000) else " Bad " ,axis=1)
```

```python
df
```

```python
def qualities(x):
    if (x["likes"]>500) and (x["comments"]<100):
        return "Risk High"
    elif x["likes"]>2000:
        return "Medium"
    elif x["shares"]>1000:
        return "Low"
```

```python
df["Qualities"]=df.apply(qualities,axis=1)
```

```python
df
```

```python
def more(x,grater,v=11):
    return x+grater+v
```

```python
df["likes"].apply(more,args=(5,),v=10)
```
