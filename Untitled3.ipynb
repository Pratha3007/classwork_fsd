```python
q1=auto.mpg.quantile(0.25)
q1
```

```python
q2=auto.mpg.quantile(0.5)
q2
```

```python
q3=auto.mpg.quantile(0.75)
q3
```

```python
auto.mpg.describe()
```

```python
iqr=q3-q1
iqr
```

```python
higher_limit=q3+1.5*iqr
lower_limit=q1-1.5*iqr
higher_limit,lower_limit
```

```python
without_outlier=auto[(auto["mpg"]<higher_limit) & (auto["mpg"]>lower_limit)]
without_outlier
```

```python
sns.boxenplot(without_outlier["mpg"])
```
